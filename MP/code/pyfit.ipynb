```python
def Nchunks(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

```python
def ChunksOfN(seq, N):

    out = []


    for i in range(math.floor(len(seq)/N)):
        out.append(seq[i*N:i*N+N])

    return out
```

```python
def log_inits(iden):
    time_series = data[data["ID"] == iden]
    N0 = min(time_series['PopBio'])
    Nmax = max(time_series['PopBio'])
    
    ts = ChunksOfN(time_series["Time"],4)
    Ns = ChunksOfN(time_series["PopBio"],4)
    #plt.scatter(x = time_series["Time"], y= time_series["PopBio"])
    
    maxgrad = 0
    rmax=0
    for t,N in zip(ts, Ns):
        res = stats.linregress(t,N)
        if res[0] > maxgrad:
            maxgrad = res[0]
            rmax = res[0] / N.iloc[2]*(1-N.iloc[2]/Nmax)
            
        t=np.array(t)
        #plt.plot(t, res[0]*t+res[1])
        
    res = stats.linregress(time_series["Time"],time_series["PopBio"])
    if res[0] > maxgrad:
        rmax = res[0] / time_series["PopBio"].iloc[-1]*(1-time_series["PopBio"].iloc[-1]/Nmax)
    return {"N0":N0, "Nmax":Nmax, "rmax":rmax}
```

```python
def gompertz_inits(iden):
    
    time_series = data[data["ID"] == iden]    
    # want tangent of log curve
    
    N0 = min(time_series['PopBio'][time_series['PopBio']>0])
    Nmax = max(time_series['PopBio'])
    A = math.log(Nmax / N0)
    time_series["PopBio"] = np.log(time_series["PopBio"]/N0)
    
    
    
    ts = ChunksOfN(time_series["Time"],4)
    Ns = ChunksOfN(time_series["PopBio"],4)
    #plt.scatter(x = time_series["Time"], y= time_series["PopBio"])
    maxgrad=0
    maxgrad = stats.linregress(time_series["Time"],time_series["PopBio"])[0]
    c=0
    i=-1
    for t,N in zip(ts, Ns):
        i+=1
        res = stats.linregress(t,N)
        if res[0] > maxgrad:
            maxgrad = res[0]
            c = res[1]
            imax = i
        t=np.array(t)
        #plt.plot(t, res[0]*t+res[1])
        
        

    #plt.plot(time_series["Time"], maxgrad*time_series["Time"]+c)
    tlag = 0.05
    if maxgrad >0 and -c > 0:
        tlag = -c / maxgrad
    #plt.ylim([min(time_series["PopBio"])*0.7,1.3*max(time_series["PopBio"])])
        
    return {"A":A, "rmax":maxgrad, "tlag":tlag}
```

```python
def baranyi_inits(iden):
    time_series = data[data["ID"] == iden]
    N0 = min(time_series['PopBio'][time_series['PopBio']>0])
    Nmax = max(time_series['PopBio'])
    
    time_series["PopBio"] = np.log(time_series["PopBio"])
    
    
    ts = ChunksOfN(time_series["Time"],4)
    Ns = ChunksOfN(time_series["PopBio"],4)
    #plt.scatter(x = time_series["Time"], y= time_series["PopBio"])
    
    maxgrad = stats.linregress(time_series["Time"],time_series["PopBio"])[1]
    c=0
    i=-1
    for t,N in zip(ts, Ns):
        i+=1
        res = stats.linregress(t,N)
        if res[0] > maxgrad:
            maxgrad = res[0]
            c = res[1]
            imax = i
        t=np.array(t)
        #plt.plot(t, res[0]*t+res[1])
        
    #plt.plot(time_series["Time"], maxgrad*time_series["Time"]+c)
    tlag = 0.05
    
    if maxgrad >0 and -c > 0:
        tlag = -c / maxgrad
        h0 = (1/((math.e**(maxgrad*tlag)) -1))
    elif maxgrad != 0:
        h0 = (1/((math.e**(maxgrad*tlag)) -1))
    else:
        h0 = (1/(math.e**(2*(math.log(Nmax)/max(time_series["Time"]))*tlag)-1))
    #plt.ylim([min(time_series["PopBio"]),max(time_series["PopBio"])])

    
    return {"N0":N0,"Nmax":Nmax,"rmax":maxgrad,"h0":h0}
```

```python
def flatten(l):
    flatter = []
    flat = True
    for item in l:
        if type(item) == list:
            flat = False
            for subitem in item:
                flatter.append(subitem)
        else:
            flatter.append(item)
    if flat:
        return flatter
    return flatten(flatter)
```

```python
sp = list(set(data.loc[:,"Species"]))
sp.sort()
combos = {}

for s in sp:
    d1 = data[data["Species"] == s]
    ts = set(d1["Temp"])
    combos[s]={}
    for t in ts:
        d2 = d1[d1["Temp"] == t]
        ms = set(d2["Medium"])
        combos[s][t] = {}
        for m in ms:
            d3 = d2[d2["Medium"] == m]
            rs = list(set(d3["Rep"]))
            combos[s][t][m] = rs

```

```python
cs=["#e41a1c",
"#377eb8",
"#4daf4a",
"#984ea3",
"#ff7f00",]
for s in combos.keys():
    d = data[data["Species"] == s]
    temps = list(combos[s].keys())
    temps.sort()
    meds = []
    for t in temps:
        for m in combos[s][t].keys():
            meds.append(m)
    meds = flatten(meds)
    meds = list(set(meds))
    meds.sort()
    medsmap={m:i for i,m in enumerate(meds)}
    tempsmap={t:i for i,t in enumerate(temps)}
    figs,axs = plt.subplots(nrows=len(temps), ncols=len(meds), figsize=[20,20])
    axs=axs.reshape([len(temps),len(meds)])
    for t in temps:
        dt = d[d["Temp"] == t]
        for m in combos[s][t].keys():
            dtm = dt[dt["Medium"] == m]
            for r in set(dtm["Rep"]):
                dtmr = dtm[dtm["Rep"] == r]
                x=dtmr["Time"]
                y=dtmr["PopBio"]
                x, y = zip(*sorted(zip(x, y)))
                axs[tempsmap[t],medsmap[m]].plot(x, y, c = cs[r-1])
                
            axs[tempsmap[t],medsmap[m]].set_title("{}_{}".format(t,m))
    figs.text(0.5, 0.04, 'Time / {}'.format(list(set(d["Time_units"]))[0]), ha='center')
    figs.text(0.04, 0.5, 'Biomass / {}'.format(list(set(d["PopBio_units"]))[0]), va='center', rotation='vertical')
#     ax.set_xlabel('Time / {}'.format(d["Time_units"][0]))
#     ax.set_ylabel('Biomass / {}'.format(d["PopBio_units"][0]))
    plt.savefig("../results/{}_plots.pdf".format(s))
```
